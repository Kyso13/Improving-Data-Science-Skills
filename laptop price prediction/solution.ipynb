```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
lr=LinearRegression()
```

```python
x=df.drop("Price",axis=1)
```

```python
y=df["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
model=lr.fit(x_train,y_train)
```

```python
pre=model.predict(x_test)
```

```python
r2_score(y_test,pre)
```

```python
mean_absolute_error(y_test,pre)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
modeldt=dt.fit(x_train,y_train)
```

```python
pre1=modeldt.predict(x_test)
```

```python
r2_score(pre1,y_test)
```

```python
mean_absolute_error(pre1,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rt=RandomForestRegressor()
```

```python
modelrt=rt.fit(x_train,y_train)
```

```python
pre2=modelrt.predict(x_test)
```

```python
r2_score(y_test,pre2)
```

```python
mean_absolute_error(y_test,pre2)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
modelad=ad.fit(x_train,y_train)
```

```python
pre3=modelad.predict(x_test)
```

```python
r2_score(y_test,pre3)
```

```python
mean_absolute_error(pre3,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
grid_search = GridSearchCV(
    estimator=modelrt, 
    param_grid=param_grid,
    cv=3,  # 3-fold cross-validation
    scoring='neg_mean_squared_error',  # Scoring metric
    verbose=2,  # Print progress
    n_jobs=-1  # Use all available cores
)
```

```python
mod=grid_search.fit(x_train, y_train)
```

```python
pre5=mod.predict(x_test)
```

```python
r2_score(pre5,y_test)
```

```python
mean_absolute_error(pre5,y_test)
```
