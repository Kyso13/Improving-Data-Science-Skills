# Yusuf Classification Project

```python
import pandas as pd

df = pd.read_csv("titanic_yusuf.csv")
pd.set_option("display.max_rows", 100)
df.head(10)
```

```python
df.columns = [i for i in range(len(df.columns))]
```

```python
df.head(100)
```

```python
def yusuf(column_index):
    last_column = df.columns[-1]
    sozluk_column = {}
    for i in df.columns:
        for j in range(df[i].nunique()):
            last_value = df.loc[df[column_index] == i, last_column]
            values = list(last_value)
            print(values)
            maximum_list = [values.count(j) for j in range(df[last_column].nunique())]
            maximum = max(maximum_list)
            sozluk_column[j] = [maximum_list.index(maximum), maximum]
    return sozluk_column
```

```python
sozlukler = [yusuf(column_index) for column_index in range(len(df.columns))]
```

```python
sozlukler
```

```python
gercek_degerler = list(df[df.columns[-1]])
print(gercek_degerler)
```

```python
df = df.drop(df.columns[-1], axis=1)
df
```

```python
yusuf_classifier = []
for index, row in df.iterrows():
    matrixler = []
    for i in range(len(df.columns)):
        print(row)
        matrixler.append(sozlukler[i][row[i]])
        max_liste = [0, 0]
    for m in matrixler:
        max_liste[m[0]] += m[1]
    yusuf_classifier.append(max_liste.index(max(max_liste)))
```

```python
a = 0
for i in range(len(yusuf_classifier)):
    if yusuf_classifier[i] == gercek_degerler[i]:
        a += 1   
print(a / len(df))
```
